# Attention Plots for Example Sources
### With individual colorbars

```python
cmap_name='magma'
normalisations = ['softmax','sigmoid','std_mean_norm','range_norm']
aggregations = ['mean','concat','deep_sup','ft']

fr1_array = np.zeros((150*4,150*4))
fr2_array = np.zeros((150*4,150*4))

# FRI Example Source Array
fig, axs = plt.subplots(4, 4, figsize=(16, 16))
for idx, norm in enumerate(reversed(normalisations)):
    for idy, agg in enumerate(aggregations):
        amap, original = attentions_func(sources[0], nets[(3-idx)*4+idy], mean=True, device=device, layer_name_base='attention', layer_no=3)
        am = amap.squeeze()
        am = am-am.min()
        am = am/am.max()
        fr1_array[150*idx:150*(idx+1),150*idy:150*(idy+1)]+=am
        ax = axs[idx, idy]
        #img = ax.imshow(amaps[idx].squeeze(), cmap='gnuplot')
        img = ax.imshow(amap.squeeze(), cmap=cmap_name)
        fig.colorbar(img, ax=ax)
        ax.set_title(f'{norm} & {agg}')
        ax.set_xticks([])
        ax.set_yticks([])
plt.show()

# FRII
fig, axs = plt.subplots(4, 4, figsize=(16, 16))
for idx, norm in enumerate(reversed(normalisations)):
    for idy, agg in enumerate(aggregations):
        amap, original = attentions_func(sources[1], nets[(3-idx)*4+idy], mean=True, device=device, layer_name_base='attention', layer_no=3)
        am = amap.squeeze()
        am = am-am.min()
        am = am/am.max()
        fr2_array[150*idx:150*(idx+1),150*idy:150*(idy+1)]+=am
        ax = axs[idx, idy]
        #img = ax.imshow(amaps[idx].squeeze(), cmap='gnuplot')
        img = ax.imshow(amap.squeeze(), cmap=cmap_name)
        fig.colorbar(img, ax=ax)
        ax.set_title(f'{norm} & {agg}')
        ax.set_xticks([])
        ax.set_yticks([])
plt.show()
```
