```python
# Out
# Print Out
# Single Example
    # Prediction
    # Attention (Individual and mean)
    # Grad CAM (Individual and mean)

def mask_on_image(img, mask):
    cm = plt.get_cmap('gist_rainbow')
    heatmap = cm(mask)
    heatmap = np.float32(heatmap)[:,:,:3]
    out = heatmap+np.float32(img)
    out = out / np.max(out)
    out_image = Image.fromarray((out[:, :, :3] * 255).astype(np.uint8))
    return out_image


for i in range(5):
    source = np.stack([fri[0,0],fri[0,0],fri[0,0]]).transpose(1,2,0) # Greyscale!
    source = source/2+0.5

    attention_map = np.stack([fri_attention[0],fri_attention[0],fri_attention[0]]).transpose(1,2,0) # Greyscale!
    overlay = mask_on_image(source,fri_attention[0]/fri_attention[0].max())

    plt.figure(figsize=(16,9))

    plt.subplot(131)
    plt.imshow(source)
    plt.contour(source[:,:,0], levels = [0.01, 0.1, 0.5], alpha=0.7)

    plt.subplot(132)
    plt.imshow(attention_map/attention_map.max())
    plt.contour(attention_map[:,:,0], levels = [0.01, 0.1, 0.5], alpha=0.7)

    plt.subplot(133)
    plt.imshow(overlay)
    plt.contour(attention_map[:,:,0], levels = [0.01], alpha=0.7)
    plt.show()
```
