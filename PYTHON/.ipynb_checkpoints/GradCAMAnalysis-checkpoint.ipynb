```python
# Extract testdata into (sorted) loopable dataset
transformation_number = 1
images = []
labels = []
data = ['MiraBest','FRDEEP-F']
data_name = data[0]
for j in range(transformation_number):
    testdata, outdata = data_call(data_name, dataloader=False)
    
    labels += outdata.targets
    for i in range(testdata.data.shape[0]):
        images.append(np.asarray(outdata[i][0]).squeeze())
        
images = np.asarray(images)/2+0.5
print(f'Images: {images.shape}')
print(f'Labels: {len(labels)}')

if data_name == 'MiraBest':
    fri = images[(0 <= np.asarray(labels)) &(np.asarray(labels)<5)]
    frii = images[(5 <= np.asarray(labels)) &(np.asarray(labels)<8)]
    hybrid = images[(8 <= np.asarray(labels))]
elif data_name == 'FRDEEP-F':
    fri = images[np.asarray(labels)==0]
    frii = images[np.asarray(labels)==1]
    hybrid = np.zeros_like(fri[:1])
```

```python
# GradCAM loop over data with best network
model_path = f'0429-FRDEEPAdadelta-playgroundv2_ft-0.0001'
#model_path = '0423-MiraBestNOHYBRIDSGD-playgroundv2_ft-0.0001'
attention_layers = ['compatibility_score1', 'compatibility_score2', 'conv6']
#model = load_net(path_to_model(model_path),)

i = 0
masks = []
gbs = []
cam_gbs = []
for img in images:
    print(np.stack([img,img,img]).shape)
    for target in attention_layers:
        img_pil = Image.fromarray(img*255).convert('RGB')
        cam_img, mask, gb, cam_gb = grad_cam_main('test/temporary.png', 
                                        use_cuda=False, 
                                        model_path=model_path, 
                                        target_layer_names=[target], 
                                        save_to='test/', 
                                        target_index=None, 
                                        save=False)
        masks.append(mask)
        gbs.append(gb)
        cam_gbs.append(cam_gb)
        if (i%10==0) & (i<157):
            plt.figure(figsize=(16,9))
            plt.subplot(151)
            plt.title(f'Image {i}: {target}')
            plt.imshow(img)
            plt.subplot(152)
            plt.title(f'CAM (raw)')
            plt.imshow(mask)
            plt.subplot(153)
            plt.title(f'GB')
            plt.imshow(gb)
            plt.subplot(154)
            plt.title(f'GradCAM')
            plt.imshow(cam_gb)
            plt.show()
    i += 1
    break
    
# plot with [original, gb, cam, cam_gb]x[compatibility_score1, compatibility_score2, conv6]

# plot sums in same format (one for each class)
```

```python
# For attention mapped networks, currently only supports a single model call. I.e. one of: ['compatibility_score1','compatibility_score2','conv6']

aggregation_modes = ['ft','deep_sup','mean','concat']

save_to = 'test/'
source_name = 'test_source'
for i in aggregation_modes:
    print(f'Starting: {i}')
    model_path = f'0423-MiraBestNOHYBRIDSGD-playgroundv2_{i}-0.0001'
    cam, gb, cam_gb = grad_cam_main(f'test/{source_name}.png', 
                                    use_cuda=False, 
                                    model_path=model_path, 
                                    target_layer_names=['compatibility_score1'], 
                                    save_to=save_to, 
                                    target_index=None, 
                                    save=True)
    plt.imshow(cam_gb)
    plt.title(f'Agg. Mode {i}')
    plt.show()
```

```python
gb_np = np.asarray(gb)
cam_np = np.asarray(cam)
cam_gb_np = np.asarray(cam_gb)
print(f'gb:\tshape:{gb_np.shape}\tmax:{gb_np.max()}\tmin:{gb_np.min()}')
print(f'cam:\tshape:{cam_np.shape}\tmax:{cam_np.max()}\tmin:{cam_np.min()}')
print(f'cam_gb:\tshape:{cam_gb_np.shape}\tmax:{cam_gb_np.max()}\tmin:{cam_gb_np.min()}')


plt.figure(figsize=(16,9))
plt.subplot(131)
plt.imshow(gb)
plt.title('Guided Backprop')
plt.subplot(132)
plt.imshow(cam)
plt.title('Class Activation Mapping')
plt.subplot(133)
plt.imshow(cam_gb)
plt.title('Grad-CAM')
plt.tight_layout()
plt.show()
```
