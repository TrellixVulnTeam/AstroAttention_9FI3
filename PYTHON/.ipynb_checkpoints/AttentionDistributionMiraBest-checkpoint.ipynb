```python
PATH = '0429-MiraBestNOHYBRIDAdadelta-playgroundv2_ft-0.0001'
#PATH = '0423-MiraBestNOHYBRIDSGD-playgroundv2_mean-0.0001'
name = 'MiraBest'
dataset = 'automatic'
model_name = path_to_model(PATH)
validation_path = 'TrainingLosses/playgroundv2_ft-0427-MiraBestNOHYBRIDAdam-0.0001_30Epochs.pt_Losses.npy.npz'
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
net = load_net(model_name,device)
```

```python
fri, frii, hybrid, images, labels = SortedDataSamples('MiraBest', 
                                                      transformed = True, 
                                                      rotations = 32, 
                                                      subset = 'NOHYBRID')
fri = np.expand_dims(fri, axis=1)
frii = np.expand_dims(frii, axis=1)
hybrid = np.expand_dims(hybrid, axis=1)
print(fri.shape)
```

```python
print(fri.shape,frii.shape,hybrid.shape)
print(images.shape)

def full_plot(source):
    plt.imshow(source[0,0])
    plt.show()
full_plot(fri[0:1])
```

```python
fri_attention, fri_attention_original = attentions_func(fri, net, device=torch.device('cuda'))
frii_attention, frii_attention_original = attentions_func(frii, net, device=torch.device('cuda'))
```

```python
# Want to do a stacking of attentions based on class:
plt.subplot(121)
plt.title('FRI Attention')
plt.imshow(fri_attention.sum(axis=0))
plt.subplot(122)
plt.imshow(fri_attention_original.sum(axis=0))
plt.colorbar()
plt.show()

plt.subplot(121)
plt.title('FRII Attention')
plt.imshow(frii_attention.sum(axis=0))
plt.subplot(122)
plt.imshow(frii_attention_original.sum(axis=0))
plt.colorbar()
plt.show()


plt.plot(fri_attention_original.sum(axis=(0,1)))
plt.plot(frii_attention_original.sum(axis=(0,1)))
plt.plot(fri_attention_original.sum(axis=(0,2)))
plt.plot(frii_attention_original.sum(axis=(0,2)))
plt.legend(['fri(1)','frii(1)','fri(2)','frii(2)'])
plt.show()
```

```python
# Want to do a stacking of attentions based on class:
plt.subplot(121)
plt.title('FRI Attention (Logged)')
plt.imshow(np.log(fri_attention.sum(axis=0)))
plt.subplot(122)
plt.imshow(np.log(fri_attention_original.sum(axis=0)))
plt.colorbar()
plt.show()

plt.subplot(121)
plt.title('FRII Attention (Logged)')
plt.imshow(np.log(frii_attention.sum(axis=0)))
plt.subplot(122)
plt.imshow(np.log(frii_attention_original.sum(axis=0)))
plt.colorbar()
plt.show()
```
