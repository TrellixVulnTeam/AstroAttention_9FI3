## Consider the Sources:

```python
# Data Selection
data = data_call('MiraBest', dataloader=False)
outset = data[1].data.squeeze()
classes = ['FRI','FRII']

# Selected due to obvious classification of respective sources
fri_np = outset[8]
frii_np= outset[92]

# Show the sources
plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(fri_np, cmap='Greys_r') #Certain FRI Example
plt.contour(fri_np)
plt.title(f'Certain FRI in Test Set')
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(frii_np, cmap = 'Greys_r') #Certain FRII Example
plt.contour(frii_np)
plt.title(f'Certain FRII in Test Set')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
# Transform images into valid input (must be tensor of given size on same device as model etc.)
fri = torch.from_numpy(fri_np ).view(1,1,150,150).float().to(device)
frii= torch.from_numpy(frii_np).view(1,1,150,150).float().to(device)
sources = [fri, frii]
```

```python
# Plot the Attention Maps that the network uses to classify

fig, axs = plt.subplots(1, len(sources), figsize=(12.8,9))
amaps = []
for idx, source in enumerate(sources):
    amap, original = attentions_func(source, net, mean=True, device=device)
    amaps.append(amap)
    img = mask_on_image(source, amap, cmap_name='gnuplot')
    axs[idx].imshow(img)
    axs[idx].set_title(f'Certain {classes[idx]}')
    axs[idx].set_xticks([])
    axs[idx].set_yticks([])
plt.show()
```
